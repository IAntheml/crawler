# 处理缺失值

## 填充缺失值方法
- 均值(mean)
- 众数(mode)
- 中位数(median)

## 众数填充

```python
# 众数存在多个
mode = df['产品数量'].mode()[0]
df['产品数量'] = df['产品数量'].fillna(value=mode)
```

## 频数最高的“类别”填充

```python
most = df['类别'].mode()[0]
# most = df['类别'].value_counts().idxmax()
# most = df['类别'].value_counts().argmax() 返回原始索引
df['类别'] = df['类别'].fillna(value=most)
```

## 向前/向后填充

```python
# method: bfill, ffill
# 向前填充ffill
df['类别'] = df['类别'].fillna(method='ffill')
# 向前填充bfill
df['类别'] = df['类别'].fillna(method='bfill')
```

## 删除缺失值

```python
# 删除存在空值的行,整表判断
df.dropna()
```

```python
# 删除指定列存在空值的行,其中一列存在空值则删除整行
df.dropna(how='any', subset=['产品数量', '类别'])
```

```python
# 删除指定列存在空值的行,均存在空值则删除整行
df.dropna(how='all', subset=['产品数量', '类别'])
```

```python
# 删除小于4个非空值的行
df.dropna(thresh=4)
```

## 字段分列
- pd.Series.str.split()

```python
# pd.Series.str.split()
df[['性别', '年龄']] = df['用户信息'].str.split(pat='，', expand=True)
df[['性别', '年龄', '用户信息']]
```

## 字段合并
- pd.Series.str.cat()

```python
# pd.Series.str.cat()
df['基本信息'] = df['性别'] + '，' + df['年龄']
# df['基本信息'] = df['性别'].str.cat(df['年龄'], '，')
df['基本信息']
```

## 内容替换
- pd.Series.str.replace()

```python
# pd.Series.str.replace()
df['交易金额'] = df['交易金额'].str.replace('元', '')
df['交易金额'] = df['交易金额'].astype(int)
df['交易金额']
```

## 信息提取
- pd.Series.str.extract()

```python
# pd.Series.str.extract()
df['手机号码'] = df['联系方式'].str.extract('(\d{11})')
df['手机号码']
```

## 爆炸序列
- 一行拆分成多行

```python
tb = df[['用户ID', '产品清单']].copy()
# split(得到列表)+explode(要求列表格式)
tb['产品清单'] = tb['产品清单'].str.split('，')
tb = tb.explode(column='产品清单', ignore_index=True)
tb
```

## 文本聚合
- 多行合并成一行

```python
# lst = ['a', 'b', 'c']
# '，'.join(lst)
tb.groupby(by='用户ID').agg(lambda x: '，'.join(x)).reset_index()
```
