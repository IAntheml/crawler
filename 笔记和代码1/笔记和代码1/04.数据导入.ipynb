# 导入Excel

```python
df1 = pd.read_excel('./data/demo_04.xlsx')
df1.head(5)
```

# 导入Excel的指定Sheet

```python
df2 = pd.read_excel('./data/demo_04.xlsx', sheet_name=1)
# df2 = pd.read_excel('./data/demo_04.xlsx',sheet_name='工作表2')
df2.head(5)
```

# 设置第2行为列索引

```python
df3 = pd.read_excel('./data/demo_04.xlsx', sheet_name=2, header=1)
df3.head(5)
```

# 通过指定的列索引来导入第1列

```python
df3 = pd.read_excel(
    './data/demo_04.xlsx',
    sheet_name=2,
    header=1,
    usecols=[0]
)
df3.head(5)
```

# 通过指定的列索引来导入第1列、第4列

```python
df3 = pd.read_excel(
    './data/demo_04.xlsx',
    sheet_name=2,
    header=1,
    usecols=[0, 3]
)
df3.head(5)
```

# 通过指定的列名来导入指定列

```python
df3 = pd.read_excel(
    './data/demo_04.xlsx',
    sheet_name=2,
    header=1,
    usecols=['username', 'mail']
)
df3.head(5)
```
