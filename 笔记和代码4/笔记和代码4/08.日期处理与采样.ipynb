# 重采样
- 针对日期字段分组聚合的快捷方式
- 重采样：升采样，降采样
- 升采样：低频率到高频率采样
- 降采样：高频率到低频率采样
- https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#dateoffset-objects
- 时间频率：
- AS: 年初
- A: 年末
- YS: 年初
- Y: 年末
- QS: 季初
- Q: 季末
- MS: 月初
- M: 月末
- W: 周
- D: 日
- H: 时
- T: 分
- min: 分
- S: 秒
- W-SUN: 周（周日结束）Sunday
- W-MON: 周（周一结束）Monday
- W-TUE: 周（周二结束）Tuesday
- W-WED: 周（周三结束）Wednesday
- W-THU: 周（周四结束）Thursday
- W-FRI: 周（周五结束）Friday
- W-SAT: 周（周六结束）Saturday

# sample函数
- 随机抽样

```python
tb = pd.DataFrame(
    data=[['貂蝉', 4],
          ['西施', 3],
          ['王昭君', 2],
          ['杨玉环', 1]],
    columns=['贵妃', '权重'])
```

```python
tb
```

## 随机抽取1个

```python
# n, 抽样数量
# frac, 抽样比例
# replace, 是否重复抽样
# weights, 抽样权重/概率
# random_state, 随机种子
# axis, 指定轴
tb.sample(n=1)
```

## 随机抽取50%

```python
tb.sample(frac=0.5)
```

## 允许重复抽样

```python
tb.sample(n=2, replace=True)
```

## 按权重抽样

```python
tb.sample(n=1, weights='权重')

# df1 = pd.DataFrame()
# for i in range(1000):
#     df2 = tb.sample(n=1, weights='权重')
#     df1 = df1.append(df2)
# df1['贵妃'].value_counts()
```
