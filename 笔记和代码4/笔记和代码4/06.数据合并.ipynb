```python
import pandas as pd
```

# merge横向合并

```python
df1 = pd.read_excel('./data/demo_06.xlsx', sheet_name='基本信息')
df2 = pd.read_excel('./data/demo_06.xlsx', sheet_name='联系方式')
```

```python
df1
```

```python
df2
```

## 左连接

```python
pd.merge(df1, df2, how='left', left_on='工号', right_on='编号')
```

## 右连接

```python
pd.merge(df1, df2, how='right', left_on='工号', right_on='编号')
```

## 外连接

```python
pd.merge(df1, df2, how='outer', left_on='工号', right_on='编号')
```

## 内连接

```python
pd.merge(df1, df2, how='inner', left_on='工号', right_on='编号')
```

## 交叉连接
- 笛卡尔积

```python
pd.merge(df1, df2, how='cross').head()
```

# join横向合并

```python
df1 = pd.read_excel('./data/demo_06.xlsx', sheet_name='基本信息')
df2 = pd.read_excel('./data/demo_06.xlsx', sheet_name='联系方式')
```

## 左连接

```python
# 将两表的关联字段设置为行索引
df1.set_index('工号', inplace=True)
df2.set_index('编号', inplace=True)
```

```python
df1.join(df2, how='left')
```

## 右连接

```python
df1.join(df2, how='right')
```

## 外连接

```python
df1.join(df2, how='outer')
```

## 内连接

```python
df1.join(df2, how='inner')
```

# concat横向合并

```python
df1 = pd.read_excel('./data/demo_06.xlsx', sheet_name='基本信息')
df2 = pd.read_excel('./data/demo_06.xlsx', sheet_name='联系方式')
```

```python
# 将两表的关联字段设置为行索引
df1.set_index('工号', inplace=True)
df2.set_index('编号', inplace=True)
```

## 外连接

```python
pd.concat(objs=[df1, df2], axis=1, join='outer')
```

## 内连接

```python
pd.concat(objs=[df1, df2], axis=1, join='inner')
```

# concat纵向合并

```python
df3 = pd.read_excel('./data/demo_06.xlsx', sheet_name='1月订单')
df4 = pd.read_excel('./data/demo_06.xlsx', sheet_name='2月订单')
```

```python
pd.concat(objs=[df3, df4], axis=0, ignore_index=True)
```

# 总结

- 表横向合并, merge方法(支持交叉连接, 也支持链式写法)
- 表横向合并, join方法(不支持交叉连接, 只支持链式写法)
- 表纵向合并, concat方法(横向合并只支持inner和outer)
