```python
import pandas as pd
```

# 导入数据

```python
df = pd.read_excel('./data/demo_09.xlsx')
```

```python
df.head()
```

# 按一列排序

```python
df.sort_values(by='销量', ascending=False)
```

# 按多列排序
- 排序方式相同，均为升序或降序

```python
df.sort_values(by=['大区', '销量'], ascending=False)
```

# 按多列排序
- 排序方式不同，一列升序一列降序

```python
df.sort_values(by=['大区', '销量'], ascending=[True, False])
```

# 对分组统计结果排序

```python
tb = df.groupby('大区')['销量'].sum().reset_index()
tb.sort_values(by='销量', ascending=False)

# df.groupby('大区')['销量'].sum().reset_index() \
#     .sort_values(by='销量', ascending=False)

# (df.groupby('大区')['销量'].sum()
#  .reset_index()
#  .sort_values(by='销量', ascending=False))
```

# rank方法排序
- method : {'average', 'min', 'max', 'first', 'dense'}

```python
df.head()
```

```python
# SQL无此函数,但可以加工实现
df['销量(avg)'] = df['销量'].rank(method='average', ascending=False)
# 类似SQL的rank
df['销量(min)'] = df['销量'].rank(method='min', ascending=False)
# SQL无此函数,但可以加工实现
df['销量(max)'] = df['销量'].rank(method='max', ascending=False)
# 类似SQL的row_number
df['销量(first)'] = df['销量'].rank(method='first', ascending=False)
# 类似SQL的dense_rank
df['销量(dense)'] = df['销量'].rank(method='dense', ascending=False)
```

```python
df.to_excel('rank.xlsx')
```
